#### PIE-CHART GRAPH

```python
data = pd.read_csv('Youth_Tobacco_Survey__YTS__Data.csv')
topic_desc_counts = data['TopicDesc'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(topic_desc_counts, labels=topic_desc_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Topic Description')
plt.axis('equal')
plt.show()
```

#### SCATTER GRAPH

```python
data = pd.read_csv('Youth_Tobacco_Survey__YTS__Data.csv')
smoking_data = data[data['TopicDesc'] == 'Cigarette Use (Youth)']
grouped_data = smoking_data.groupby(['YEAR', 'Age'])['Data_Value'].mean().unstack()
grouped_data.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Smoking Prevalence')
plt.title('Smoking Prevalence Among Different Age Groups Over the Years')
plt.legend(title='Age Group')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('Youth_Tobacco_Survey__YTS__Data.csv')
smoking_data = data[data['TopicDesc'] == 'Cigarette Use (Youth)']
grouped_data = smoking_data.groupby(['YEAR', 'Gender'])['Data_Value'].mean().unstack()
grouped_data.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Smoking Prevalence')
plt.title('Smoking Prevalence Among Different Genders Over the Years')
plt.legend(title='Gender')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('Youth_Tobacco_Survey__YTS__Data.csv')
smoking_data = data[data['TopicDesc'] == 'Cigarette Use (Youth)']
grouped_data = smoking_data.groupby(['YEAR', 'Education'])['Data_Value'].mean().unstack()
grouped_data.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Smoking Prevalence')
plt.title('Smoking Prevalence Among Different Education Levels Over the Years')
plt.legend(title='Education Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## THE LAST TOPIC LINE OF REGRESSION 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data = pd.read_csv('Youth_Tobacco_Survey__YTS__Data.csv')
smoking_data = data[data['TopicDesc'] == 'Cigarette Use (Youth)']
grouped_data = smoking_data.groupby('YEAR')['Data_Value'].mean().reset_index()
X = grouped_data['YEAR'].values.reshape(-1, 1)
y = grouped_data['Data_Value'].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('Year')
plt.ylabel('Smoking Prevalence')
plt.title('Linear Regression of Smoking Prevalence Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
